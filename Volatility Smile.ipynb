```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

ticker = yf.Ticker("AAPL")
expirations = ticker.options
pd.DataFrame(expirations, columns=["Expiration Dates"])
```

```python
data = ticker.history(period="5d", interval="1m") #1 min interval data for the last 5 days
data
```

```python
ticker.history(period="5d") #stocks in the last 5 days
```

```python
ticker.options[0]
```

```python
option_chain = ticker.option_chain(expirations[0]) #Choose expiration date
calls = pd.DataFrame(option_chain.calls)
puts = pd.DataFrame(option_chain.puts)
calls
```

```python
calls = calls[calls.openInterest > 800] #filtering for options with more than 800 open potitions held / open contracts
calls
```

```python
plt.figure(figsize=(12,6))
plt.plot(calls['strike'], calls['impliedVolatility'], marker='p', linestyle='-', color='b')
plt.xlabel('Strike Price')
plt.ylabel('Implied Volatility')
plt.title('Implied Volatility vs Strike Price @ Expiration Date: ' + expirations[0])
```
